# Import grades into dataframe
API: https://ubcgrades.com/api/v3/grades/UBCV/2022S
Options: `/grades/campus/yearsession` 
- campus: `UBCV` or `UBCO`
Note: '[V3] carries data from 2019W onward...Some endpoints are not available in v3, and v2 should be used instead.'

NOTE: this opens up a webapplication on localhost. For me, it is port 8050 (localhost:8050)

```python
from dash import Dash, html, dcc, callback, Output, Input
import plotly.express as px
# import pandas as pd

# df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminder_unfiltered.csv')

app = Dash(__name__)

app.layout = html.Div([
    html.H1(children='Title of Dash App', style={'textAlign':'center'}),
    dcc.Dropdown(df.faculty_title.unique(), 'Faculty of Arts', id='dropdown-selection'),
    dcc.Dropdown(df.subject.unique(), "CPSC", id="xaxis-column"),
    dcc.Dropdown(df.course.unique(), "210", id="yaxis-column"),
    dcc.Graph(id='graph-content')
    dcc.Slider(
            df.year.min(),
            df.year.max(),
            step=None,
            id="year--slider",
            value=df.year.max(),
            marks={str(year): str(year) for year in df.year.unique()},
        ),
])

@callback(
    Output('graph-content', 'figure'),
    Input('dropdown-selection', 'value')
    Input("year--slider", "value"),
)
def update_graph(value):
    dff = df[df.faculty_title==value]
    return px.line(dff, x='course', y='average')

if __name__ == '__main__':
    app.run(debug=True)
```
